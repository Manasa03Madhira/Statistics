```python
import scipy.stats as  stats
import math
import pandas as py
import numpy as np
from scipy.stats import poisson
np.random.seed(5)
school_ages=stats.poisson.rvs(loc=10,mu=32,size=100)
classA_ages=stats.poisson.rvs(loc=15,mu=30,size=100)
```

```python
classA_ages
```

```python
school_ages
```

```python
school_ages.mean()
```

```python
classA_ages.mean()
```

```python
from scipy.stats import ttest_1samp
school_age=np.array(school_ages)
school_age
_,p_value=ttest_1samp(classA_ages, popmean=school_age.mean())
p_value
```

```python
school_age.mean()
```

```python
if p_value<0.05:
    print("Accept Ho")
else:
    print("Reject Ho")
```

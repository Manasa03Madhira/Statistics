```python
#Normal distribution:
from numpy import random
df=random.normal(loc=4,scale=3,size=(2,4))
print(df)
```

```python
#loc = mean, scaale= standard deviation, size= the size of return of an array
```

```python
df=random.normal(size=(3,10))
print(df)
```

```python
#Visualization of normal distribution
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(random.normal(size=100),hist=False, kde = True)
plt.show()
```

```python
#binomial distribution
#n = number of trails, p = probability of occurence of each trial , size = the shape of returned array
df = random.binomial(n= 1000, p=0.8, size = 1000)
print(df)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show()
```

```python
#poisson distribution
#lam - rate or known number of occurrences e.g. 2 for above problem., size - The shape of the returned array.
```

```python
df=random.poisson(lam=2,size=100)
print(df)
```

```python
#Visualization for Poisson distribution
sns.distplot(random.poisson(lam=2,size=100),hist=False,kde=True)
plt.show()
```

```python
#df - (degree of freedom)., size - The shape of the returned array.
```

```python
#Chisquare 
df=random.chisquare(df=3,size=15)
print(df)
#Visualization:
sns.distplot(random.chisquare(df=1, size=15), hist=False)

plt.show()
```

```python
#pareto distribution
x=random.pareto(a=2,size=(2,3))
print(x)
```

```python
#Visualization:
sns.distplot(random.pareto(a=2, size=1000),hist=False, kde=True)
plt.show()
```
